```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Set 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
# Create a chart without annotations. 
x_axis=np.arange(0,len(gyms))
plt.bar(x_axis, members)
```

```python
x_axis=np.arange(0,len(gyms))
tick_locations=[x for x in x_axis]
plt.xticks(tick_locations, gyms)
plt.xlabel("Gym Name")
plt.ylabel("# Members")
plt.title("Gym Popularity")
plt.xlim(-0.75,len(gyms)-0.25)
plt.ylim(0, max(members)+3)

plt.bar(x_axis, members, facecolor="green")

plt.show()
```

```python
explode=(0,0,0.1,0)
plt.title("Gym Popularity")
plt.pie(members,labels=gyms,autopct="%1.1f%%", startangle=90, explode=explode)

plt.show()
```
