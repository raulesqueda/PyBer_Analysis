```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_df=pd.read_csv("Classes/got.csv")
got_df.head()
```

```python
# Get attacker and defender king data
attacker_data = got_df["attacker_king"].value_counts()
defender_data=got_df["defender_king"].value_counts()

defender_data
```

```python
# Get total battle data
battle_df=pd.DataFrame({
    "Attacker":attacker_data,
    "Defender":defender_data
})

battle_df
```

```python
# Configure plot and ticks
battle_df.plot(kind="bar")

# Set textual properties
plt.title("Bloodthrist of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battles")
plt.xticks(rotation=45)

# Show plot
plt.show()
# Resize plot to display labels
plt.tight_layout()
```

```python
battle_totals_df=attacker_data.add(defender_data, fill_value=0)
battle_totals_df
```

```python
# Configure plot and ticks
battle_totals_df.plot(kind="bar")

# Set textual properties
plt.title("Bloodthrist of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battles")
plt.xticks(rotation=45)

# Show plot
plt.show()
# Resize plot to display labels
plt.tight_layout()
```
